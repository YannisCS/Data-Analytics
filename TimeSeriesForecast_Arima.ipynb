## Custom functions

```python
def score_metrics(field, data, models):
    mae = []
    rmse = []
    mape = []

    for model in models:
        mae.append(mean_absolute_error(data[field], data[model]))
        rmse.append(mean_squared_error(data[field], data[model], squared=False))
        mape.append(mean_absolute_percentage_error(data[field], data[model]))

    result = pd.DataFrame({
        "ModelName": models,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    })

    return result
```

### 2.1 Additive Time Series Decomposition

Additive decomposition: $y_t = S_t + T_t + R_t$

```python
##-- Function used to apply decomposition and plot result --## 
def ATS_graph(data):
    decompose = seasonal_decompose(x=data,model='additive')
    trend = decompose.trend
    seasonal = decompose.seasonal
    remainder = decompose.resid
    
    fig, axes = plt.subplots(4, 1, sharex=True, sharey=False)
    fig.set_figheight(10)
    fig.set_figwidth(20)
    # First plot to the Original time series
    axes[0].plot(data, label='Original') 
    axes[0].legend(loc='upper left');
    # second plot to be for trend
    axes[1].plot(trend, label='Trend')
    axes[1].legend(loc='upper left');
    # third plot to be Seasonality component
    axes[2].plot(seasonal, label='Seasonality')
    axes[2].legend(loc='upper left');
    # last last plot to be Residual component
    axes[3].plot(remainder, label='Residuals')
    axes[3].legend(loc='upper left');
```

```python
ATS_graph(res)
```

### 2.2 Exponential Smoothing

```python
##-- Function used to apply es and return forecast test data --##
def ES(train,test):
    # Apply ES
    simple = SimpleExpSmoothing(train)
    trend = ExponentialSmoothing(train, trend="add")
    dampened = ExponentialSmoothing(train, trend="add", damped_trend=True)
    seasonal = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=7)
    
    # Test
    test["Simple"] = simple.fit().forecast(test_period)
    test["Trend"] = trend.fit().forecast(test_period)
    test["Dampened"] = dampened.fit().forecast(test_period)
    test["Seasonal"] = seasonal.fit().forecast(test_period)
```

```python
ES(res_train,res_test)
```

```python
res_test.plot()
```

This plot shows that the seasonal (purple line) is close to the original pattern (blue), let's check its scores

```python
es = ["Simple","Trend","Dampened","Seasonal"]
res_es_score = score_metrics("Residential",res_test,es)

cm = sns.color_palette("vlag", as_cmap=True)
res_es_score.style.background_gradient(cmap=cm)
```

Seasonal performs the best as we find in the plot

based on the parameters auto arima gave us, we want to double check the p and q in order by AIC/ BIC, we will call the result as ref-arima

```python
# ACF, PACF plot
def tsplot(y,lags=None,title='',figsize=(14,8)):
    '''
    import statsmodels.api as sm
    import statsmodels.formula as smf
    import statsmodels.tsa.api as smt
    import matplotlib.pylab as plt
    import seaborn as sns
    '''
    
    fig = plt.figure(figsize=figsize)
    layout=(2,2)
    ts_ax = plt.subplot2grid(layout,(0,0))
    hist_ax = plt.subplot2grid(layout,(0,1))
    acf_ax = plt.subplot2grid(layout,(1,0))
    pacf_ax = plt.subplot2grid(layout,(1,1))
    
    y.plot(ax=ts_ax)
    ts_ax.set_title(title)
    y.plot(ax=hist_ax,kind='hist',bins=25)
    hist_ax.set_title('Histogram')
    smt.graphics.plot_acf(y,lags=lags,ax=acf_ax)
    smt.graphics.plot_pacf(y,lags=lags,ax=pacf_ax)
    [ax.set_xlim(0) for ax in [acf_ax,pacf_ax]]
    sns.despine()
    fig.tight_layout()
    return ts_ax, acf_ax, pacf_ax

# bic heatmap plot
def bic(data):
    import itertools
    p_min, p_max = 0, 4
    d_min, d_max = 0, 0
    q_min, q_max = 0, 4
    result_bic = pd.DataFrame(index=['AR{}'.format(i) for i in range(p_min,p_max+1)],
                              columns=['MA{}'.format(i) for i in range(q_min, q_max+1)])
    for p,d,q in itertools.product(range(p_min, p_max+1),
                                   range(d_min, d_max+1),
                                   range(q_min, q_max+1)):
        if p==0 and d==0 and q==0:
            result_bic.loc['AR{}'.format(p),'MA{}'.format(q)]=np.nan
            continue
        try:
            model = ARIMA(data,order=(p,d,q))
            results = model.fit()
            result_bic.loc['AR{}'.format(p),'MA{}'.format(q)] = results.bic
        except:
            continue
    result_bic = result_bic[result_bic.columns].astype(float)
    fig, ax = plt.subplots(figsize=(10,8))
    ax = sns.heatmap(result_bic,
                     mask = result_bic.isnull(),
                     ax = ax,
                     annot=True,
                     fmt='.2f',
                    );
    ax.set_title('BIC')
    
# all in one function
# reutrn min order
def choose_pq(data,lags=None,title=''):
    tsplot(data,lags=lags,title=title,figsize=(14,8))
    bic(data)
    # the smaller bic or aic, the better
    res_train_results = sm.tsa.arma_order_select_ic(res_train,
                                                    ic=['aic','bic'],
                                                    trend='n',
                                                    max_ar=4,
                                                    max_ma=4)
    
    print('AIC', res_train_results.aic_min_order)
    print('BIC', res_train_results.bic_min_order)
    min_order = res_train_results.aic_min_order
    return min_order[0], min_order[1]
```

```python
res_ref_p, res_ref_q = choose_pq(res_train,title='Rresidential',lags=20)
```
